```python
import pandas as pd
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python
url ='https://www.turing.com/interview-questions/python'
```

```python
client =urlopen(url)
```

```python
html = client.read()
html
```

```python
client.close()
```

```python
soup = bs(html,'html.parser') 
soup
```

```python
containers =soup.find_all("div",{"class":"OrderedBlocks_section__SLmhw"})
```

```python
questian_py =containers[0].findAll("div",{"class": "RichText_root__TVuc9 OrderedBlocks_question__DKDfv"})
questian_py[0].text
```

```python
Answering_py = containers[0].findAll('div',{"class" :"RichText_root__TVuc9 OrderedBlocks_answerText__rgqDt"})
Answering_py[0].text
```

```python
f= open ('python_developer_interview.txt','w',encoding ='utf-8')
#header ='Questian_py, Answering_py,'
#f.write(header)
     
```

```python
for container in containers:
    Questian_py =container.findAll("div",{"class": "RichText_root__TVuc9 OrderedBlocks_question__DKDfv"})
    Questian_py =Questian_py[0].text.strip()
    
    Answering_py = container.findAll('div',{"class" :"RichText_root__TVuc9 OrderedBlocks_answerText__rgqDt"})
    Answering_py =Answering_py[0].text.strip()
    
    f.write( Questian_py + ',' + Answering_py +'\n')
f.close()
```

```python
files=["AI_for_interview.txt","python_developer_interview.txt"]
with open("Ai_pyDev.txt","w",encoding='utf-8')as f: 
    for file in files:
        with open(file,"r",encoding='utf-8')as fr:  
            for line in fr:
                f.write(line)
     
```
